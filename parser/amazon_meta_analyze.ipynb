```python
for product_index in range(len(new_get_all_index) - 1):
  product = {}
  cpidex = new_get_all_index[product_index]
  npidex = new_get_all_index[product_index + 1]
  get_partition = test_list_label[cpidex[1]: npidex[1]]
  #print(test_list_label[cpidex[1]: npidex[1]])
  #print((cpidex[1],  npidex[1]))
  #list_result.append(len(get_partition))

  #Extract ID field
  #label_id, value_id = get_partition[0].split(':', 1)
  #product = {label_id.strip().lower(): int(value_id)}

  #Extract ASIN field
  label_asin, value_asin = get_partition[1].split(':', 1)
  product[label_asin.strip().lower()] = value_asin.strip()

  #Extract Title field
  label_title, value_title = get_partition[2].split(':', 1)
  product[label_title.strip()] = value_title.strip()

  #Extract group field
  label_group, value_group = get_partition[3].split(':', 1)
  product[label_group.strip()] = value_group.strip()

  #Extract salesrank field
  label_salesrank, value_salesrank = get_partition[4].split(':', 1)
  product[label_salesrank.strip()] = int(value_salesrank)


  #Extract similar field
  label_similar, value_similar = get_partition[5].split(':', 1)
  product[label_similar.strip()] = value_similar.split()[1:]

  #Extract properties from categories and fields
  temp_list_categories = []
  label_categories, value_categories = get_partition[6].split(':', 1)
  value_categories = int(value_categories)
  if value_categories > 0:
    for subpartition_categories_elements in get_partition[7: 7 + value_categories]:
      temp_list_categories.append(subpartition_categories_elements)
      #print(subpartition_categories_elements)
  product[label_categories.strip()] = temp_list_categories

  #Extract reviews properties and field
  label_reviews, value_reviews = get_partition[7 + value_categories].split(':', 1)
  value_reviews =  value_reviews.strip().split()
  value_reviews = {
      value_reviews[0].strip()[:-1]: int(value_reviews[1]),
      value_reviews[2].strip()[:-1]: int(value_reviews[3]),
      f"{value_reviews[4].strip()} {value_reviews[5].strip()[:-1]}": float(value_reviews[6]),
      'entries': []
  }

  #temp_list_reviews = []
  if value_reviews['total'] > 0:
    for subpartition_reviews_elements in get_partition[8 + value_categories: (8 + value_categories) + value_reviews['total']]:
      subpartition_reviews_elements = subpartition_reviews_elements.strip().split()
      if len(subpartition_reviews_elements) <= 4:
        continue
      subpartition_reviews_elements = {
          'date': subpartition_reviews_elements[0].strip(),
          subpartition_reviews_elements[1].strip()[:-1]: subpartition_reviews_elements[2].strip(),
          subpartition_reviews_elements[3].strip()[:-1]: int(subpartition_reviews_elements[4]),
          subpartition_reviews_elements[5].strip()[:-1]: int(subpartition_reviews_elements[6]),
          subpartition_reviews_elements[7].strip()[:-1]: int(subpartition_reviews_elements[8])

      }
        #print(get_partition)
        #print(subpartition_reviews_elements)

      value_reviews['entries'].append(subpartition_reviews_elements)

  #product[label_reviews.strip()] = temp_list_reviews

  product[label_reviews.strip()] = value_reviews
  getfd.add(value_title.strip().lower())
  #print(value_asin.strip().lower())
  #print(get_partition)
  get_final_data.append(product)













```
